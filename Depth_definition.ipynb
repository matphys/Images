```python
import imageio
import numpy as np
from matplotlib import pyplot as plt
from skimage.io import imread, imshow, imsave
```

```python
image_names = ['2.jfif','probagavno.jfif','proba2_bad.jfif']
images = [np.array(imageio.imread(fname)).astype('uint64') for fname in image_names]
gray_images = [img.mean(axis=2).astype('uint64') for img in images]

plt.imshow(images[0])
plt.show()
plt.imshow(gray_images[0], cmap = 'gray')
plt.imshow(gray_images[1], cmap = 'gray')

for y in range(len(gray_images)): #      
    plt.imshow(gray_images[y], cmap = 'gray')
    plt.show()
```

```python
depth_maps = [np.log(gray_img.max() / gray_img) for gray_img in gray_images]
depth_maps
```

```python
max_depths = [depth_map.max() for depth_map in depth_maps]
max_depths
```
